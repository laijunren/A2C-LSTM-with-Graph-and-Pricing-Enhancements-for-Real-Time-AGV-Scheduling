```python
import matplotlib.pyplot as plt
import numpy as np

def moving_average(data, window_size):
    """ Calculate the moving average over a specific window size """
    return np.convolve(data, np.ones(window_size)/window_size, mode='valid')

# Reading the rewards from the file
file_path = 'rewards.txt'  # Replace with your actual file path
with open(file_path, 'r') as file:
    rewards = [float(line.strip()) for line in file]

# Smoothing the rewards
window_size = 100  # Adjust this for more or less smoothing
smoothed_rewards = moving_average(rewards, window_size)

# Plotting the rewards
plt.figure(figsize=(10, 6))
plt.plot(rewards, label='Original Rewards', alpha=0.5)  # Original rewards (more transparent)
plt.plot(range(window_size-1, len(rewards)), smoothed_rewards, label='Smoothed Rewards', color='orange')  # Smoothed rewards
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.title('Rewards Over Episodes (Smoothed)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Function to read loss data from a file
def read_loss_data(file_path):
    with open(file_path, 'r') as file:
        loss_data = [float(line.strip()) for line in file]
    return loss_data

# Function to smooth the data with a moving average
def smooth_data(data, window_size):
    return np.convolve(data, np.ones(window_size)/window_size, mode='valid')

# Function to plot actor and critic losses separately with smoothing
def plot_losses_separately(actor_loss, critic_loss, window_size=164):
    # Smooth the data
    smoothed_actor_loss = smooth_data(actor_loss, window_size)
    smoothed_critic_loss = smooth_data(critic_loss, window_size)

    # Plot for Actor Loss
    plt.figure(figsize=(10, 5))
    plt.plot(smoothed_actor_loss, label='Smoothed Actor Loss', color='blue')
    plt.xlabel('Training Steps')
    plt.ylabel('Actor Loss')
    plt.title('Smoothed Actor Loss Over Time')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Plot for Critic Loss
    plt.figure(figsize=(10, 5))
    plt.plot(smoothed_critic_loss, label='Smoothed Critic Loss', color='red')
    plt.xlabel('Training Steps')
    plt.ylabel('Critic Loss')
    plt.title('Smoothed Critic Loss Over Time')
    plt.legend()
    plt.grid(True)
    plt.show()

# Reading the loss data from files
actor_loss = read_loss_data('/home/aaa/my_code/hospital-main/A2C_output/actor_loss_12_10_06_17.txt')
critic_loss = read_loss_data('/home/aaa/my_code/hospital-main/A2C_output/critic_loss_12_10_06_17.txt')

# Plotting the smoothed losses separately
plot_losses_separately(actor_loss, critic_loss)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# 
file_path = '/home/aaa/my_code/hospital-main/Simulator/Hospital_DRL/output/adjacency_matrix.csv'
adj_matrix = pd.read_csv(file_path, index_col=0)  # 

# 
graph = nx.from_pandas_adjacency(adj_matrix, create_using=nx.DiGraph)

#  3D 
pos_2d = nx.spring_layout(graph, seed=42)  # 2D 
pos_3d = {node: (x, y, np.random.uniform(0, 1)) for node, (x, y) in pos_2d.items()}  #  z 

#  3D 
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# 
for node, (x, y, z) in pos_3d.items():
    ax.scatter(x, y, z, s=100, c='lightblue', edgecolors='black', alpha=0.9)
    ax.text(x, y, z, node, fontsize=10, ha='center', va='center', color='black')

# 
for start_node, end_node in graph.edges():
    x_vals = [pos_3d[start_node][0], pos_3d[end_node][0]]
    y_vals = [pos_3d[start_node][1], pos_3d[end_node][1]]
    z_vals = [pos_3d[start_node][2], pos_3d[end_node][2]]
    ax.plot(x_vals, y_vals, z_vals, color='gray', alpha=0.7)

# 
ax.set_title("3D Graph Visualization from Adjacency Matrix", fontsize=15)
ax.view_init(elev=20, azim=30)  # 

plt.show()
```
